```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv("housing.csv")
df
```

```python
df['income_cat'] = pd.cut(df['median_income'],bins=[0,1.5,3.0,4.5,6.0,np.inf], labels=[1,2,3,4,5])
df
```

```python
from sklearn.model_selection import StratifiedShuffleSplit
split = StratifiedShuffleSplit(n_splits=1,test_size=0.2,random_state=42)
for train_index,test_index in split.split(df,df['income_cat']):
    strat_train_set=df.loc[train_index]
    strat_test_set=df.loc[test_index]
```

```python
for sett in (strat_train_set,strat_test_set):
    sett.drop("income_cat", inplace=True,axis=1)
```

```python
data = strat_train_set.copy()
data
```

```python
housing = data.drop("median_house_value",axis=1)
housing_labels = data['median_house_value']
```

```python
housing
```

```python
from sklearn.pipeline import Pipeline
from sklearn.preprocessing import StandardScaler
from sklearn.impute import SimpleImputer
```

```python
housing = housing.drop("ocean_proximity",axis=1)
```

```python
mypipeline = Pipeline([
    ("impute", SimpleImputer(strategy="median")),
    ("Standarize", StandardScaler()),
])
```

```python
house = mypipeline.fit_transform(housing)
```

```python
housing = pd.DataFrame(house  , columns=housing.columns, index=housing.index)
```
