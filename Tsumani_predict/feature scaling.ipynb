```python
from sklearn.model_selection import StratifiedShuffleSplit
import numpy as np
```

```python
import pandas as pd

df = pd.read_csv("earthquake.csv")
df
```

```python
df['mmi_cat'] = pd.cut(df['mmi'] , bins = [0.0,3.0,6.0,9.0,12.0,np.inf] , labels=[1,2,3,4,5])
```

```python
df.drop('dmin_cat' , axis=1)
```

```python
split = StratifiedShuffleSplit(n_splits=2 , test_size=0.2 , random_state=42)
for train_index, test_index in split.split(df,df['mmi_cat']):
    train_set = df.iloc[train_index].drop('mmi_cat' ,axis=1)
    test_set = df.iloc[test_index].drop('mmi_cat' , axis=1)
```

```python
data = train_set.copy()
```

```python
data_features = data.drop('tsunami', axis=1)
data_labels = data['tsunami']
```

```python
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
data_feature = scaler.fit_transform(data_features)
```
